```python
import pandas as pd
import numpy as np

from src.stop_move_detection import move_detection
```

```python
# Folder from which we'll read the data.
folder_in = './data_simulator/huge_dataset/'

# Folder to which we'll save the data.
folder_out = './data_simulator/huge_dataset/'


# Read the simulator's trajectories.
path_dataset = folder_in + 'dataset_simulator_trajectories.compressed.parquet'
df_traj = pd.read_parquet(path_dataset)
display(df_traj.info())

# Read the stop segments detected over the trajectories.
stop_dataset = folder_in + 'dataset_simulator_trajectories.compressed.parquet.stops.parquet'
df_stops = pd.read_parquet(stop_dataset)
display(df_stops.info())
```

```python
# Create a parquet containing the trajectories' move segments.
df_moves = move_detection(df_traj, df_stops)
display(df_moves)

df_moves.to_parquet(folder_out + 'dataset_simulator_trajectories.compressed.parquet.moves.parquet', index = True)
```
